## Two options for subplots

```python
# Subplot options 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
```

```python
# Subplots options 1
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each different axis
ax0.plot(x, x/2);
ax1.scatter(np.random.random(10), np.random.random(10));
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=3,
                      figsize=(20, 10))
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[0, 2].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 0].barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
ax[1, 1].scatter(x, np.sin(x));
ax[1, 2].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matplotlib OO (Object Oriented) method?)

* When plotting something quickly, okay to use the pyplot method.
* When plotting something more advanced, usee the OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(heart_disease)
```

```python
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x='age',
            y='chol',  # chol is short for cholestorol
            c='target');  # c is short for color.
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45, 80])
```

```python
over_50.target.values
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line as the mean
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Customizing the y and x axis limitations

# Subplot of chol, age, thalach (max heart rate)
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(15, 15),
                              sharex=True)  # Keeping the same x axis values only on one subplot.

## Subplot 1:
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")  # colormap

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Change the x-axis limits
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="target")

# Add a meanline to ax0
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

## Subplot 2:
# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")  # colormap

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Change ax1 x-axis and y-axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="target")

# Add a meanline to ax1
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
